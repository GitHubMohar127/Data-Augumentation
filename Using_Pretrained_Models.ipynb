```python
# Import necessary libraries for Keras and ResNet50.
import keras
from keras.applications.resnet50 import ResNet50
from keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
# Load the ResNet50 model with pre-trained ImageNet weights.
model = ResNet50(weights='imagenet')
```

```python
# Define the path to the input image.
img_path = '/content/Cat.jpg'
```

```python
# Load and preprocess the image for the ResNet50 model.
img = keras.utils.load_img(img_path, target_size=(224, 224))
x = keras.utils.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
# Make predictions using the loaded ResNet50 model.
preds = model.predict(x)
```

```python
# Decode and print the top 5 predictions from the model.
print('Predicted:', decode_predictions(preds, top=5)[0])
```

```python
# Import necessary libraries for Keras and VGG16.
import keras
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
import numpy as np
```

```python
# Load the VGG16 model with pre-trained ImageNet weights, excluding the top classification layer.
model = VGG16(weights='imagenet', include_top=False)
```

```python
# Define the path to the input image and preprocess it for the VGG16 model.
img_path = '/content/Dog.jpg'
img = keras.utils.load_img(img_path, target_size=(224, 224))
x = keras.utils.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
# Extract features from the preprocessed image using the VGG16 model.
features = model.predict(x)
```
