```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/drive/MyDrive/ML/Data Augmentation/train/Cat.jpg', target_size=(200,200))
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1, 200, 200, 3)
```

```python
i=0

for output in datagen.flow(input_batch, batch_size=1, save_to_dir='/content/drive/MyDrive/ML/Data Augmentation/aug', save_prefix='cat', save_format='jpeg'):
  i+=1
  if i == 10:
    break
```
